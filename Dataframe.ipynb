## Broadcasting

```python
nba = pd.read_csv("nba.csv")
nba.head(3)
```

```python
nba["Age"]
nba["Age"].add(5) #You can add 5 to NaN value, but pandas keep NaN
nba["Age"] + 5 # Add 5 to every value in this series

nba["Salary"].sub(5000000) #subtract
nba["Salary"] - 5000000 # subtract 5m from every value

nba["Weight"].mul(0.453592)
nba["Weight in kg"] = nba["Weight"] * 0.453592 #generate a brand new series
```

## Drop rows with null value

```python
nba = pd.read_csv("nba.csv")
nba.tail(3)
```

```python
nba.dropna() #This drop all rows where there is one or more NaN value
nba.dropna(how = "all", inplace = True) # Remove rows if all of the values are NaN
# You can set inplace to True
```
