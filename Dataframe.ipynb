## The .astype() Method 

```python
import pandas as pd
nba = pd.read_csv("nba.csv").dropna(how = "all")
nba["Salary"].fillna(0, inplace = True)
nba["College"].fillna("None", inplace = True)
nba.head()
```

```python
nba.dtypes
```

```python
nba.info()
```

```python
nba["Salary"] = nba["Salary"].astype("int")  #.astype() method doesn't have intype, so we have to reassign it "Salary" series
nba.head(5)
```

```python
nba["Age"] = nba["Age"].astype("int")
nba["Number"] = nba["Number"].astype("int")
nba.head(3)
# nba.astype("float")
```

## Category data type. 
* Reduce memory usage when you have a amall number of unique values

```python
nba["Position"].value_counts()
```

```python
nba["Position"].nunique()
```

```python
nba["Position"] = nba["Position"].astype("category")
```

```python
nba["Team"].value_counts()
```

```python
nba["Team"].nunique()
```

```python
nba["Team"] = nba["Team"].astype("category")
```
