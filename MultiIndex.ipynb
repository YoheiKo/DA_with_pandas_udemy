## Create A Multiindex with the .set_index() Method

```python
bigmac = pd.read_csv("bigmac.csv", parse_dates= ["Date"])
bigmac.head(3)
```

```python
bigmac.set_index("Date").head() # Set single string as index
```

```python
bigmac.set_index(keys = "Date").head() # Set single strings as index
```

```python
bigmac.set_index(keys = "Country").head() # Set single strings as index
```

```python
bigmac.set_index(keys = ["Date", "Country"])
# A multiindex is an index with multi labels. Two or more values are need to identify a row
# "Date" is the first level index here.
# Same date exist for many countries
```

* Column that has least number of unique value should serve as outer layer / parent-level index
* In this case, it is "Date"

```python
bigmac.set_index(keys = ["Country", "Date"]).head(10) # you can reverse the order 
```

## The .get_level_values() Method

```python
import pandas as pd
bigmac = pd.read_csv("bigmac.csv", parse_dates= ["Date"], index_col = ["Date", "Country"])
bigmac.sort_index(inplace = True) # pandas works faster in a sorted order. This is the best practice. 
bigmac.head()
```

```python
bigmac.index # This is a object. We can call methods on it
```

```python
bigmac.index.get_level_values(0) #First-level index
```

```python
bigmac.index.get_level_values(1)  #Second-level index
```

```python
bigmac.index.get_level_values("Date")
```

```python
bigmac.index.get_level_values("Country")
```

## The .set_names() Method on MultiIndex

```python
bigmac = pd.read_csv("bigmac.csv", parse_dates= ["Date"], index_col = ["Date", "Country"])
bigmac.sort_index(inplace = True)
bigmac.head()
```

```python
bigmac.index.set_names(["Day", "Location"], inplace = True) 
# Provide full new list to set_names()
# "Date" is going to change to "Day". "Country" is going to change to "Location"
```

```python
bigmac.head(10)
```

```python
bigmac.sort_index(ascending = [True, False]).head(10) #Day is going to be ascending, and Location is going to be descending order.
```

## Extract Rows from a MultiIndex Data

```python
bigmac = pd.read_csv("bigmac.csv", parse_dates= ["Date"], index_col = ["Date", "Country"])
bigmac.sort_index(inplace = True)
bigmac.head()
```

```python
bigmac.loc["2010-01-01"] .head()
# loc is a method that takes square brackets.
```

```python
bigmac.loc[("2010-01-01")].head()
# loc is a method that takes square brackets.
```

```python
bigmac.loc[("2010-01-01", "Brazil")].head() 
# We have to pass a turple instead of a list in a multiindex DataFrame. 
# loc is a method that takes square brackets.
```

```python
bigmac.loc[("2010-01-01", "Brazil"), "Price in US Dollars"]
# 1st argument of loc() is the horizontal index. Second argument is the columns 
```

```python
bigmac.loc[("2015-07-01")].head()
```

```python
bigmac.loc[("2015-07-01", "Chile")]
```

```python
bigmac.loc[("2015-07-01", "Chile"), "Price in US Dollars"] 
# specify the columns in the second argument
# Pass turple as the first argument. Pass index in the order that they appear in the original DataFrame.
```

## The .transpose() Method 

```python
bigmac = pd.read_csv("bigmac.csv", parse_dates= ["Date"], index_col = ["Date", "Country"])
bigmac.sort_index(inplace = True)
bigmac.head()
```

```python
bigmac.transpose()  # .transpose() swap the axsis. 
#This method does not affect the original DataFrame. And it does not have inplace parameter. 
```

```python
bigmac = bigmac.transpose()
bigmac
```

```python
bigmac.loc["Price in US Dollars"].head(10) #!st argument is the row-level label/labels, 2nd is the column-level label/labels
```

```python
bigmac.loc["Price in US Dollars", ("2016-01-01")].head(10)
```

```python
bigmac.loc["Price in US Dollars", ("2016-01-01", "Canada")]
```

## The .swaplevel() Method

```python
bigmac = pd.read_csv("bigmac.csv", parse_dates= ["Date"], index_col = ["Date", "Country"])
bigmac.sort_index(inplace = True)
bigmac.head()
```

```python
bigmac = bigmac.reset_index() # This is one way to swap multi index 
bigmac.head()
```

```python
bigmac.set_index(keys = ["Country", "Date"], inplace = True) # This is one way to swap multi index "(cont)"
bigmac.head()
```

```python
bigmac = pd.read_csv("bigmac.csv", parse_dates= ["Date"], index_col = ["Date", "Country"])
bigmac.sort_index(inplace = True)
bigmac.head()
```

```python
bigmac = bigmac.swaplevel().head()
bigmac.head()
```

## The .unstack() Method, part

```python
world = pd.read_csv("worldstats.csv", index_col = ["country", "year"])
world.head()
```

```python
s = world.stack()
s.head()
```

```python
s.unstack().head() # .unstack() takes the deepest index/layer and move it to columns axis
```

```python
s.unstack().unstack().head(10) # .unstack() takes the deepest index/layer and move it to columns axis
```

## The .unstack() Method, Part2

```python
world = pd.read_csv("worldstats.csv", index_col = ["country", "year"])
world.head()
```

```python
s = world.stack()
s.head()
```

```python
s.unstack(level = 2).head()
# Outer layer "country" is going to have a index position of zero. 2nd layer "Year" is going to have a index position of one
```

```python
s.unstack(level = 0).head() 
```

```python
s.unstack(level = 1).head() 
```

```python
s.unstack(level = -1).head()   # In this DataFrame, this is idenical to s.unstack(level = 2).head()
```

```python
s.unstack(level = -2).head() # Identical to s.unstack(level = 1).head() 
```

```python
s.unstack(level = -3).head() # Identical to s.unstack(level = 0).head() 
```

```python
s.head()
```

```python
s.unstack("year").head() # You can specify the layer by its name
```
