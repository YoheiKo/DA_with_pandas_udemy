## Filtering with the where method

```python
import pandas as pd
bond = pd.read_csv("jamesbond.csv", index_col = "Film")
bond.sort_index(inplace = True)
bond.head()
```

```python
mask = bond["Actor"] == "Sean Connery"
bond[mask] # Returns small subset DataFrame
```

```python
bond.where(mask) # Entire row values are going to be present where mask is met. All other rows are NaNs.
```

```python
bond.where(bond["Box Office"] > 800)
```

```python
mask2 = bond["Box Office"] > 800
```

```python
bond.where(mask & mask2)
```

## A review of the .apply() Method on Single Columns

```python
import pandas as pd
bond = pd.read_csv("jamesbond.csv", index_col = "Film")
bond.sort_index(inplace = True)
bond.head()
```

```python
def cts_m(num):
    return str(num) + " M"
```

```python
bond["Box Office"] = bond["Box Office"].apply(cts_m) 
# apply method takes existing python metod and applies it to every values in the Series
bond.head()
```

```python
bond["Budget"] = bond["Budget"].apply(cts_m)
bond.head()
```

```python
import pandas as pd
bond = pd.read_csv("jamesbond.csv", index_col = "Film")
bond.sort_index(inplace = True)
bond.head()
```

```python
columns = ["Box Office", "Budget", "Bond Actor Salary"]
for col in columns: # col is a temporary variable
    bond[col] = bond[col].apply(cts_m)
bond.head()
```

## The .apply() Method with Row Values

```python
import pandas as pd
bond = pd.read_csv("jamesbond.csv", index_col = "Film")
bond.sort_index(inplace = True)
bond.head()
```

```python
def godm(row): # iterate over each row
    actor = row[1]
    budjet = row[4]
    
    if actor == "Pierce Brosnan":
        return "The Best"
    elif actor == "Roger Moore" and budjet > 40:
        return "Enjoyable"
    else:
        return "I have no clue"
```

```python
bond.apply(godm, axis = "columns") # axis = "columns" -  for each row you move across columns.
```

### If you want to work on a copy fo the original Dataframe, you have to copy it

```python
import pandas as pd
bond = pd.read_csv("jamesbond.csv", index_col = "Film")
bond.sort_index(inplace = True)
bond.head()
```

```python
directors = bond["Director"].copy() # This is a new isoloated Series
directors.head(3)
```

```python
directors["A View to a Kill"] = "Mr. John Glen"
```

```python
directors.head(3) # Value is overwritten. But tbhis is a separate copy.
```

```python
bond.head(3) # The value is remained on the original Dataframe
```
