## The .set_index() and .reset_index() Methods

```python
import pandas as pd
```

```python
#bond = pd.read_csv("jamesbond.csv", index_col= "Film") #index_col set the column name you like to serve as an index
bond = pd.read_csv("jamesbond.csv")
bond.head(3)
```

```python
bond.set_index("Film", inplace = True) # move film from the regular column to the index on the left side
bond.head(5)
```

```python
bond.reset_index() #reset the index to an index starting from zero
bond.head(5)
```

```python
bond.reset_index(drop = False, inplace = True) #reset the index to an index starting from zero
bond.head(10)
```

```python
bond.set_index("Film", inplace = True)
bond.head(10)
```

### We have to reset the index before setting a new index.
#### Otherwise our current index will be wiped out

```python
bond.reset_index(inplace = True)
bond.set_index("Year", inplace = True)
bond.head(10)
```

## Retrieve Rows by Index Label with .loc[]

```python
import pandas as pd
bond = pd.read_csv("jamesbond.csv", index_col = "Film")
bond.sort_index(inplace = True) # sorting an index can greatly accelerate the extraction process when the data is large
bond.head()
```

```python
bond.loc["Goldfinger"] # .loc method take square brackets. 
```

```python
bond.loc["GoldenEye"]
```

```python
bond.loc["Casino Royale"] # Unlike python dictionary, pandas do not have to have strict unique values in the index 
```

```python
bond.loc["Diamonds Are Forever" : "Moonraker"] #when working with string, last value is inclusive, Not the case for numerical index 
```

```python
bond.loc["GoldenEye": ]
```

```python
bond.loc[: "You Only Live Twice"]
```

```python
bond.loc[["Moonraker", "Octopussy"]]  # extract multiple non-sequential row
```

```python
bond.loc[["Octopussy", "Moonraker"]] #extract in a specific order
```

```python
bond.loc[["For Your Eyes Only", "Live and Let Die", "Gold Bond"]] #extract a row that doesn't exist int the original df. You have to be careful
```

```python
"Gold Bond" in bond.index
```

```python
"Live and Let Die" in bond.index
```

## Retrieve Row(s) by Index Position with .iloc method
iloc is short for index location

```python
bond = pd.read_csv("jamesbond.csv")
bond.head()
```

```python
bond.loc[15]
```

```python
bond.iloc[15]
```

```python
bond.iloc[[15,20]]
```

```python
bond.iloc[:4]  # When we use numerical position, the end is exclusive 
```

```python
bond = pd.read_csv("jamesbond.csv", index_col = "Film")
bond.sort_index(inplace = True)
bond.head()
```

```python
bond.iloc[0] # Behind the scenes each of the rows has specific numerical index position
```

```python
bond.iloc[:5] # 5 is exclusive
```

```python
bond[18:]
```

```python
bond.iloc[[2,4,6,8]]
```
