## Change Options with Attributes and Dot Syntax

```python
import numpy as np
import pandas as pd
```

```python
data = np.random.randint(0, 100, [1000, 50]) 
df = pd.DataFrame(data)
df.tail(2)
# generate random integers between 0 and 100. The last attribute is the matrix size.
```

```python
pd.options.display.max_rows # get the values
```

```python
pd.options.display.max_columns
```

```python
df
```

```python
pd.options.display.max_rows = 24 # set the max_rows values to 20
```

```python
pd.options.display.max_columns = 24
```

```python
df
```

## Change pandas Options with Methods

```python
data = np.random.randint(0, 100, [1000, 50])
df = pd.DataFrame(data)
df.head()
```

```python
pd.get_option("max_rows")
```

```python
pd.get_option("max_columns")
```

```python
pd.set_option("max_rows", 48)
```

```python
pd.get_option("max_rows")
```

```python
pd.options.display.max_rows # work interchangeably
```

```python
pd.reset_option("max_rows") # reverse to default value
```

```python
pd.get_option("max_rows")
```

```python
pd.reset_option("max_columns")
```

```python
pd.get_option("max_columns")
```

```python
pd.describe_option("max_rows")  # load documentation
```

## The precision Option

```python
df = pd.DataFrame(np.random.randn(5,5)) # attributes are dimension of dataframe 
pd.reset_option("precision")
df
# by default there are six degits after the decimal point
```

```python
pd.get_option("precision")
```

```python
pd.set_option("precision",3)
# You can preserve the pricision and modify the output
```

```python
pd.get_option("precision")
```

```python
df
```

```python
pd.reset_option("precision")
df
```
