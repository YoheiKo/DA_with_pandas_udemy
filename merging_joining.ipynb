## The pd.concat() Method, Part1

```python
week1 = pd.read_csv("Restaurant - Week 1 Sales.csv")
week2 = pd.read_csv("Restaurant - Week 2 Sales.csv")
customers = pd.read_csv("Restaurant - Customers.csv")
foods = pd.read_csv("Restaurant - Foods.csv")
```

```python
len(week1)
```

```python
len(week2)
```

```python
pd.concat([week1, week2]).tail() # We have two pairs of index ranging from 0 to 249
```

```python
pd.concat([week1, week2], ignore_index = True).tail() 
# "ignore_index = True" ignore the present index and generate a new index
```

```python
sales = pd.concat([week1, week2], keys = ["week1", "week2"]) # keys argument retains the orignal data source information
```

```python
sales.head()
```

```python
sales.loc["week1"].head()
```

```python
sales.loc[("week1",0)]
```

```python
sales.loc[("week1",0), "Customer ID"]
```

## Inner Joins, Part1
* logical AND

```python
week1 = pd.read_csv("Restaurant - Week 1 Sales.csv")
week2 = pd.read_csv("Restaurant - Week 2 Sales.csv")
customers = pd.read_csv("Restaurant - Customers.csv")
foods = pd.read_csv("Restaurant - Foods.csv")
```

```python
week1.merge(week2, how = "inner", on = "Customer ID") 
# merge is a method. merge is called on the left-side DataFrame
# how is the way of merging
# on is basically specify what column or columns you want to merge. Columns names have to be exact same name.
```
