## Outer Joins
* Either or Both
* Either but not Both
* Both DataFrame have equal weight on this operation

```python
import pandas as pd
week1 = pd.read_csv("Restaurant - Week 1 Sales.csv")
week2 = pd.read_csv("Restaurant - Week 2 Sales.csv")
customers = pd.read_csv("Restaurant - Customers.csv")
foods = pd.read_csv("Restaurant - Foods.csv")
```

```python
week1.head()
```

```python
week2.head()
```

```python
week1.merge(week2, how = "outer", on = "Customer ID", suffixes = ["_w1", "_w2"]).head(10)
# Whereever we no NaN value in a row, that means that Customer ID was found in both of the DataFrame.
# Customer 537 came in the first week and bought food item #9.0 and came in the second week and bought food item #5.0.
# Whereever we have NaN, that indicate that that value does not exist in either one of our DataFrames.
# Customer 97 came in the first week and bought food item #4.0 but did not come in the week2.
# Pandas need something to store in the empty cell. It provides NaN in it.
```

```python
len(week1.merge(week2, how = "outer", on = "Customer ID", suffixes = ["_w1", "_w2"]))
```

```python
week1[week1["Customer ID"] == 537]
```

```python
week2[week2["Customer ID"] == 537]
```

```python
merged = week1.merge(week2, how = "outer", on = "Customer ID", suffixes = ["_w1", "_w2"], indicator = True)
merged.head(10)
```

```python
merged["_merge"].value_counts() 
# 197 unique customers exist in right DataFrame/week2 and 195 unique costmers exist in the left DataFrame/week1. 
```

## Choose either

```python
mask = merged["_merge"].isin(["left_only", "right_only"])
merged[mask].head()
```

## Left Joins - equivalent to excel Vlookup
* Set left DataFrame as our main fous. 
* Look at the right DataFrame from the left and if there are common value, pull the rows from the right into the left.
* We don't care about the unique values in the right.
* Right Joins is the swap of Left Joins

```python
import pandas as pd
week1 = pd.read_csv("Restaurant - Week 1 Sales.csv")
week2 = pd.read_csv("Restaurant - Week 2 Sales.csv")
customers = pd.read_csv("Restaurant - Customers.csv")
foods = pd.read_csv("Restaurant - Foods.csv")
```

```python
week1.head()
```

```python
foods
```

```python
week1.merge(foods, how = "left", on = "Food ID").head(10)
```

```python
week1.merge(foods, how = "left", on = "Food ID", sort = True).head(10) 
# sort by matching columns. A short cut of sort_value method
```

## The left_on and right_on Parameters

```python
import pandas as pd
week1 = pd.read_csv("Restaurant - Week 1 Sales.csv")
week2 = pd.read_csv("Restaurant - Week 2 Sales.csv")
customers = pd.read_csv("Restaurant - Customers.csv")
foods = pd.read_csv("Restaurant - Foods.csv")
```

```python
week2.head()
```

```python
customers.head() 
# The name of the matching column is "ID", different from "Customer ID".
# "on" parameter is no longer going to work
```

```python
week2.merge(customers, how = "left", left_on = "Customer ID", right_on = "ID").head()
# Both IDs are dragged in because they have different name. 
```

```python
week2 = week2.merge(customers, how = "left", left_on = "Customer ID", right_on = "ID", sort = True).drop("ID", axis = "columns").head()
# Drop duplicate extra column
# sort by Costomer ID
```

```python
week2.head()
```

## Merging by Indexes with left_index and right_index Parameters

```python
import pandas as pd
week1 = pd.read_csv("Restaurant - Week 1 Sales.csv")
week2 = pd.read_csv("Restaurant - Week 2 Sales.csv")
customers = pd.read_csv("Restaurant - Customers.csv", index_col = "ID")
foods = pd.read_csv("Restaurant - Foods.csv", index_col = "Food ID")
```

```python
customers.head(3)
```

```python
foods.head(3)
```

```python
week1.head(3)
```

```python
sales = week1.merge(customers, how = "left", left_on = "Customer ID", right_index = True)
sales.head()
# If we want to merge DataFrame on Index right_on and left_on will not work
# Duplicate columns do not arise 
```

```python
len(week1)
```

```python
len(sales)
```

```python
sales2 = sales.merge(foods, how = "left", left_on = "Food ID", right_index = True)
sales2.head()
```
