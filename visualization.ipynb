## Bar Charts

```python
companies = ["GOOG"]

goog = data.DataReader(name = companies, data_source = "yahoo", start = "2004-01-01", end = "2018-12-31")
goog.head()
```

```python
def rank_performance(stock_price):
    if stock_price <= 200:
        return "Poor"
    elif stock_price > 200 and stock_price <= 500:
        return "Satisfactory"
    else:
        return "Stellar"
```

```python
rank_performance(600)
```

```python
goog["Close"].squeeze() # .squeeze is the invert of to_frame
```

```python
plt.style.use("ggplot")
goog["Close"].squeeze().apply(rank_performance).value_counts().plot(kind = "bar")
# .squeeze is the invert of to_frame
# .apply call function on every sngle value
# .plot(kind = "bar") plots a bar graph
```

```python
plt.style.use("ggplot")
goog["Close"].squeeze().apply(rank_performance).value_counts().plot(kind = "barh")
# .squeeze is the invert of to_frame
# .apply call function on every sngle value
# .plot(kind = "bar") plots a bar graph
```

## Pie Charts

```python
companies = ["AAPL"]

apple = data.DataReader(name = companies, data_source = "yahoo", start = "2012-01-01", end = "2018-12-31")
apple.head()
```

```python
apple["Close"].mean()
```

```python
def rank_performance(stock_price):
    if stock_price >= 112.241198:
        return "Above average"
    else:
        return "Below average"
```

```python
apple["Close"].squeeze().apply(rank_performance).value_counts().plot(kind = "pie")
```

## Histogram

```python
companies = ["GOOG"]

goog = data.DataReader(name = companies, data_source = "yahoo", start = "2004-01-01", end = "2018-12-31")
goog.head()
```

```python
def custom_round(price):
    return int(price / 100.00) * 100 # rounding down
```

```python
custom_round(193)
```

```python
googlist = goog["Close"].squeeze()
```

```python
plt.style.use("ggplot")
plt.hist(googlist)
```
